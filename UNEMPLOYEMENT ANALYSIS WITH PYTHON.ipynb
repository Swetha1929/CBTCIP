```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the datasets
unemployment_2020 = pd.read_excel("Unemployment_Rate_upto_11_2020.xlsx")
unemployment_india = pd.read_excel("Unemployment in India.xlsx")
```

```python
# Display basic information about the datasets
print("Unemployment data up to November 2020:")
print(unemployment_2020.head())
```

```python
print("\nUnemployment data in India:")
print(unemployment_india.head())
```

```python
# Check for missing values
print("\nMissing values in Unemployment data up to November 2020:")
print(unemployment_2020.isnull().sum())

print("\nMissing values in Unemployment data in India:")
print(unemployment_india.isnull().sum())
```

```python
# Plotting unemployment trends over time
plt.figure(figsize=(10, 6))
plt.plot(unemployment_2020['Date'], unemployment_2020['Estimated Unemployment Rate (%)'], label='Up to Nov 2020')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Trend')
plt.legend()
plt.show()
```

```python
# Descriptive statistics of Unemployment Rate
print("\nDescriptive statistics of Unemployment Rate up to November 2020:")
print(unemployment_2020['Estimated Unemployment Rate (%)'].describe())

print("\nDescriptive statistics of Unemployment Rate in India:")
print(unemployment_india['Estimated Unemployment Rate (%)'].describe())
```
